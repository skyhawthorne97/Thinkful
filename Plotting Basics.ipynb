Plotting Basics

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Cleaned_Youth_Behavior_Risk_Survey.csv')
print(df)
```

```python
smoking = [df.iloc[0,3], df.iloc[7,3]]
plt.plot(smoking)
plt.title("Smoking Rate")
plt.xlabel("Year")
plt.ylabel("Prevalence")
plt.show
```

```python
plt.plot(df.iloc[ :5,3], color='purple')
plt.ylabel('Prevalence')
plt.title('2011 vs 2013 Rates')
plt.plot(df.iloc[5: ,3], color='green')
plt.show()
```

```python
plt.plot(df.iloc[4:9:4, 3], color = 'purple')
plt.plot(df.iloc[5:10:4, 3], color = 'green')
plt.title('The Correlation of Physical Activity and Obesity')
plt.show()
```

```python
plt.hist(df['Prevalence'])

plt.show()
```

```python
plt.boxplot(df['Prevalence'])
plt.show()
```

```python
plt.scatter(x = df['Prevalence'], y = df['Year'], color='green')
plt.show()
```
